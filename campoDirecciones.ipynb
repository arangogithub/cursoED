# Campos de direcciones

Se genera el campo de direcciones de una ecuación diferencial de la forma
$$
\frac{dx}{dt} = f(t,x)
$$
en una cuadrícula de $M\times N$ puntos en el rectángulo   $[t_0, t_0+ Lt]\times[x_0, x_0+ Lx].$ En cada punto $(t,x)$ de la cuadrícula se traza un segmento de pendiente $f(t,x)$ y longitud $l$. Por simplicidad se eleige $(t,x)$ como punto de inicio del segmento.   

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Parámetros

```python
t0, Lt = 0, 2
x0, Lx = 0,2
M, N = 15, 15
l = 0.1
```

## Cuadrícula

```python
tvalores = np.linspace(t0, t0+Lt, N)
xvalores = np.linspace(x0, x0+Lx, M)
tt, xx =  np.meshgrid(tvalores,xvalores)
```

```python
fig = plt.figure(figsize =(5,5))
ax = fig.subplots(1, 1)
ax.scatter(tt,xx, s=12, c='r', alpha=1)
plt.show()
```

## Campo de direcciones

```python
# Define la función f(t,x)
f = lambda t,x : t-x
```

```python
def trazaSegmento(t,x):
    '''
    Traza un segmento de recta longitud l que empieza 
    en (t,x) y tiene pendiente f(t,x)
    '''
    h = l/np.sqrt(1+f(t,x)**2)
    tf = t+h
    xf = x + h*f(t,x)
    return plt.plot([t,tf], [x,xf ],alpha=0.5, c = 'b', lw = 0.75 )  
```

```python
fig = plt.figure(figsize =(5,5))
ax = fig.subplots(1, 1)
ax.scatter(tt,xx, s=12, c='r', alpha=0.5)
for t in tt:
    for x in xx:
        trazaSegmento(t,x)
plt.show()
```

## Soluciones

```python
# Soluciones de la ecuación diferencial
sol = lambda c,t : c*np.exp(-t)+ t -1
```

```python
sol = lambda c,t : c*np.exp(-t)+ t -1
t_dominio = np.linspace(t0, t0+Lt, 100)
sol1_t = sol(1,t_dominio)
sol2_t = sol(2,t_dominio)
sol3_t = sol(3,t_dominio)
```

```python
fig = plt.figure(figsize =(5,5))
ax = fig.subplots(1, 1)
ax.plot(t_dominio, sol1_t, lw=5, c='r', alpha=0.25)
ax.plot(t_dominio, sol2_t, lw=5, c='r', alpha=0.25)
ax.plot(t_dominio, sol3_t, lw=5, c='r', alpha=0.25)
for t in tt:
    for x in xx:
        trazaSegmento(t,x)
plt.show()
```
